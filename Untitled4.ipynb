```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load dataset (after download or from Excel)
df = pd.read_excel("Mushroom_WITH_Duplicates.xlsx")  # file from UCI or generated

# Step 2: Check for duplicates
print("Duplicates found:", df.duplicated().sum())

# Step 3: Remove duplicates
df = df.drop_duplicates()

# Step 4: Check for missing values
print("Missing values:\n", df.isnull().sum())

# Step 5: Encode target column: 'class' (e = 0, p = 1)
df['class'] = df['class'].map({'e': 0, 'p': 1})

# Step 6: Label encode all other categorical columns
label_encoders = {}
for column in df.columns:
    if df[column].dtype == 'object':
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
        label_encoders[column] = le

# Step 7: Save cleaned dataset
df.to_excel("Mushroom_Cleaned.xlsx", index=False)

# Step 8: Display first few rows of the final dataset
print(df.head())

# Step 9: Plot class distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=df, palette='Set2')
plt.title('Mushroom Class Distribution (0 = Edible, 1 = Poisonous)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks([0, 1], ['Edible (0)', 'Poisonous (1)'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
